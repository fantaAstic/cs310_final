```python
from collections import Counter
import pandas as pd
from collections import Counter
import ast
import os
import google.generativeai as genai

def compute_module_id(module_name: str) -> int: # should return a number, can be a random number tbh
    """Generate a unique ID for a module based on its name."""
    pass

def compute_summary(df: pd.DataFrame, module_name: str) -> str:
    """Generate a summary of reviews, sentiment trends, and emotions for a given module."""
    # Step 1: Group by 'month' and calculate the average emotion score for each month
    curr_module_reviews = df[df['name'] == module_name].copy()  # Use .copy() to avoid warning

    # Assuming 'emotion_label' contains categorical labels (not scores), we might need to adjust the aggregation
    # Here we can count the occurrences of each emotion per month
    monthly_emotion_counts = curr_module_reviews.groupby(['month', 'emotion_label']).size().unstack(fill_value=0)
    monthly_emotion_trend = monthly_emotion_counts.diff(axis=0).max(axis=1)  # This finds if there's any increase in emotion counts

    # Find the last month where emotion count improved
    last_emotion_improvement_month = monthly_emotion_trend[monthly_emotion_trend > 0].idxmax()

    # If it's a string, convert it to datetime
    if isinstance(last_emotion_improvement_month, str):
        last_emotion_improvement_month = pd.to_datetime(last_emotion_improvement_month, format='%Y-%m')

    # Step 2: Group by 'month' and calculate the average rating for each month
    monthly_avg_rating = curr_module_reviews.groupby('month')['rating'].mean()
    monthly_avg_rating_diff = monthly_avg_rating.diff()

    # Find the last month where rating improved
    last_rating_improvement_month = monthly_avg_rating_diff[monthly_avg_rating_diff > 0].idxmax()

    # If it's a string, convert it to datetime
    if isinstance(last_rating_improvement_month, str):
        last_rating_improvement_month = pd.to_datetime(last_rating_improvement_month, format='%Y-%m')

    # Step 3: Flatten the lists from the 'topic_summary' column into one long list
    all_topics = [topic for sublist in curr_module_reviews['all_topics'] for topic in sublist]
    topic_counts = Counter(all_topics)
    most_common_topic = topic_counts.most_common(1)[0]

    # Step 4: Get the most recent month
    curr_module_reviews.loc[:, 'month'] = pd.to_datetime(curr_module_reviews['month'], format='%Y-%m')  # Use .loc to modify safely
    most_recent_month = curr_module_reviews['month'].max()

    # Step 5: Find the most popular sentiment and emotion for the most recent month
    recent_month_reviews = curr_module_reviews[curr_module_reviews['month'] == most_recent_month]
    most_recent_sentiment = recent_month_reviews['sentiment_label_from_review'].value_counts().idxmax()
    most_recent_emotion = recent_month_reviews['emotion_label'].value_counts().idxmax()

    # Step 6: Count consecutive months where the sentiment and emotion stayed the same
    same_rating_months = count_consecutive_months(curr_module_reviews, most_recent_month, 'rating')
    same_emotion_months = count_consecutive_months(curr_module_reviews, most_recent_month, 'emotion_label')

    # Prepare the final outlook string
    summary_str = f'''
    Most reviews are {most_recent_sentiment} and about {most_common_topic[0]} with most students feeling {most_recent_emotion}.
    There has been no improvement in sentiment since {last_rating_improvement_month.strftime('%Y-%m')} and the last improvement in emotion
    was in {last_emotion_improvement_month.strftime('%Y-%m')}.
    The most recent reviews are {most_recent_sentiment} and {most_recent_emotion}.
    Over the existence of the course, the number of {most_recent_sentiment} and {most_recent_emotion} reviews has increased.
    The sentiment has stayed the same for {same_rating_months} consecutive month(s) and the emotion has stayed the same for
    {same_emotion_months} consecutive month(s).
    '''

    # Return or print the final summary string
    return summary_str



def count_consecutive_months(df, most_recent_month, column):
    # Filter data from the most recent month going backward
    df_sorted = df[df['month'] <= most_recent_month].sort_values('month', ascending=False)

    # Get the most recent value for the column
    recent_value = df_sorted[df_sorted['month'] == most_recent_month][column].values[0]

    # Count consecutive months where the value stays the same
    count = 0
    for value in df_sorted[column]:
        if value == recent_value:
            count += 1
        else:
            break
    return count

def compute_outlook(df: pd.DataFrame, module_name: str) -> str:
    """Determine the outlook (e.g., 'positive', 'negative', 'mixed') for a module based on sentiment trends."""

    # Filter the reviews for the specified module
    module_reviews = df[df['name'] == module_name]

    # Calculate the percentage of positive reviews (assuming 'positive' is the label for positive sentiment)
    positive_reviews = module_reviews[module_reviews['sentiment_label_from_review'] == 'positive']
    total_reviews = len(module_reviews)

    # Calculate percentage
    positive_percentage = (len(positive_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    # Determine the outlook based on the percentage of positive reviews
    if positive_percentage > 70:
        outlook = 'Positive'
    elif positive_percentage < 50:
        outlook = 'Negative'
    else:
        outlook = 'Mixed'

    return f"The outlook for '{module_name}' is: {outlook}"


def compute_positive_reviews(df: pd.DataFrame, module_name: str) -> str:
    """Calculate the percentage of positive reviews for a module."""
    # Filter the reviews for the specified module
    module_reviews = df[df['name'] == module_name]

    # Count positive reviews (Assume 'positive' is the label for positive reviews)
    positive_reviews = module_reviews[module_reviews['sentiment_label_from_review'] == 'positive']
    total_reviews = len(module_reviews)

    # Calculate percentage
    positive_percentage = (len(positive_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    # Return the result as a formatted string
    return f"Percentage of positive reviews for '{module_name}': {positive_percentage:.2f}%"

def compute_negative_reviews(df: pd.DataFrame, module_name: str) -> str:
    """Calculate the percentage of negative reviews for a module."""
    # Filter the reviews for the specified module
    module_reviews = df[df['name'] == module_name]

    # Count negative reviews (Assume 'negative' is the label for negative reviews)
    negative_reviews = module_reviews[module_reviews['sentiment_label_from_review'] == 'negative']
    total_reviews = len(module_reviews)

    # Calculate percentage
    negative_percentage = (len(negative_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    # Return the result as a formatted string
    return f"Percentage of negative reviews for '{module_name}': {negative_percentage:.2f}%"

def get_category(module_name: str) -> str:
    """Assign a category to a module based on its name."""
    pass

def get_topics(module_name: str) -> str:
    """Assign a category to a module based on its name."""
    pass

def compute_teacher_feedback_prompt(df: pd.DataFrame, module_name: str) -> str:
    """Generate a feedback prompt for a given module based on the feedback summary in the dataset."""

    # Filter the data for the specific module
    module_reviews = df[df['name'] == module_name]

    # Assuming 'rating' represents overall feedback and sentiment column is 'sentiment_label_from_review'
    # Adjust these as per your actual dataset columns
    avg_rating = module_reviews['rating'].mean()
    sentiment_percentage = (module_reviews['sentiment_label_from_review'] == 'positive').mean() * 100

    # Get the most common emotions (you can adjust how to retrieve emotions)
    # For example, assuming 'emotion_label' contains emotion data
    most_common_emotions = module_reviews['emotion_label'].mode().head(3).tolist()

    # Get the main areas of commentary from 'topic_subtopic_dict' column
    # Ensure it's properly converted from string to dictionary
    subtopics = ast.literal_eval(module_reviews['topic_subtopic_dict'].iloc[0])  # Convert string to dictionary

    # Format the commentary for each subtopic
    formatted_commentary = "\n".join([f"{key}: {value}" for key, value in subtopics.items()])

    # For time trend, assuming you have 'month' column (if not, adjust based on your dataset)
    # You could calculate sentiment or rating trends over months, for simplicity, let's assume positive trend:
    # Example logic for time trend improvement (adjust based on actual data)
    time_trend = "Positive feedback consistently increased throughout the semester, especially after midterm."

    outlook = compute_outlook(df, module_name)

    # Construct the prompt
    prompt = f'''
    I’m a teacher of a '{module_name}' module, and I recently received feedback with the following summary:

    Overall: {outlook} feedback ({avg_rating:.1f} average rating, {sentiment_percentage:.1f}% positive sentiment).

    Key (most common) Emotions: {', '.join(most_common_emotions)}.

    Main Areas of Commentary:
    {formatted_commentary}

    Time Trend: {time_trend}

    I need guidance on how to build on this positive feedback and further enhance the student experience. Specifically, I'd like recommendations that are:

    Specific & Actionable: For each area of positive commentary (e.g., Exam, Assignment, Lecturer, and Grading), provide me with practical, specific actions I can take to maintain or improve these areas.
    Preventative: How can I prevent complacency and continue improving these areas in future semesters or iterations of the module?
    Emotionally Sensitive: How can I acknowledge the positive feedback and use it to motivate students further, while maintaining a positive learning environment?
    Prioritised: If there are multiple things I could do to improve, which areas should I focus on first and why?
    Considered for Time Trend: How can I use the positive time trend to ensure that future semesters continue on this upward trajectory?

    I’m looking for a response that is practical, forward-thinking, and specific to my situation as a teacher for a '{module_name}' module.
    '''

    return prompt

def compute_teacher_feedback_recommendation(df: pd.DataFrame, module_name: str) -> str:
    """Generate recommendations for teachers based on the feedback analysis for a module."""
    genai.configure(api_key='SECRET')

    # Create the model
    generation_config = {
      "temperature": 0.7, # need to check temp param
      "top_p": 0.95,
      "top_k": 40,
      "max_output_tokens": 8192,
      "response_mime_type": "text/plain",
    }

    model = genai.GenerativeModel(
      model_name="gemini-2.0-flash-exp",
      generation_config=generation_config,
    )

    chat_session = model.start_chat(
      history=[
        {
          "role": "user",
          "parts": [
            "something to give recommendations to teachers to improve modules based on student feedback on modules that they teach\n",
          ],
        },
        {
          "role": "model",
          "parts": [
            "Okay, let's craft something that can provide teachers with actionable recommendations based on student feedback on their modules. Here's a breakdown of how it could work, along with examples:\n\n**Core Components:**\n\n1. **Feedback Collection:**\n   * **Structured Surveys:** This is the most common method. Questions could cover:\n      * **Clarity of Content:** (e.g., Was the material presented in a clear and understandable way?)\n      * **Engagement:** (e.g., Did you find the module engaging? Did it hold your interest?)\n      * **Organization:** (e.g., Was the module well-organized? Did the flow of topics make sense?)\n      * **Usefulness:** (e.g., How relevant was the material to your learning goals?)\n      * **Pacing:** (e.g., Was the pace of the module appropriate?)\n      * **Assessment:** (e.g., Were the assessments fair and aligned with the content?)\n      * **Instructor's Role:** (e.g., Was the instructor responsive and helpful?)\n      * **Overall Experience:** (e.g., How satisfied were you with the module overall?)\n   * **Open-Ended Feedback:** Allow students to provide written comments on what they liked, what they didn't, and suggestions for improvement.\n   * **Anonymous Feedback:** Encourage honest responses by ensuring anonymity.\n\n2. **Data Analysis:**\n   * **Quantitative Analysis:**\n      * **Average Scores:** Calculate average scores for each question to identify areas of strength and weakness.\n      * **Frequency Distributions:** Analyze how often students selected specific responses (e.g., how many students rated a question as \"Excellent\").\n      * **Trend Analysis:** Track changes in scores over time to monitor improvement or decline in different areas.\n   * **Qualitative Analysis:**\n      * **Theme Identification:** Look for common themes and patterns in the open-ended feedback.\n      * **Sentiment Analysis:** Use tools to analyze the overall tone (positive, negative, neutral) of the comments.\n\n3. **Recommendation Generation:** This is the key part. Here's how to generate meaningful recommendations:\n   * **Based on Quantitative Data:**\n      * **Low-Scoring Areas:** If a question consistently scores low, suggest specific actions (e.g., \"The question 'Clarity of Content' scored below average. Consider providing more detailed examples or explanations.\")\n      * **Inconsistent Ratings:** If there's a wide spread of ratings, this might indicate that the content appeals to some but not others. Recommend offering optional extensions or different approaches to the material.\n   * **Based on Qualitative Data:**\n       * **Recurring Themes:** If multiple students mention that the pacing is too fast, recommend slowing down. If multiple students mention that the instructor's responsiveness is good, highlight this strength.\n       * **Specific Suggestions:** If a student suggests adding more real-world examples, suggest incorporating them.\n   * **Focus on Actionable Advice:** Recommendations should be specific and easily translated into concrete changes. Avoid vague statements.\n   * **Provide Context:** Present data in a way that is clear and understandable for the teacher, along with comparisons to previous years or other similar modules (if available).\n   * **Highlight Strengths:**  Don't focus solely on weaknesses.  Also include positive feedback to reinforce what is working well.\n\n**Example Scenario and Recommendations:**\n\n**Scenario:**\n\n* **Module:** \"Introduction to Programming\"\n* **Feedback:**\n   * **Quantitative:**  Low average score on the \"Clarity of Content\" question.\n   * **Qualitative:**  Multiple students commented that \"the concepts were confusing\" and \"the examples were too abstract.\"\n\n**Generated Recommendations:**\n\n* **Recommendation 1 (Quantitative):**  \"The average score for 'Clarity of Content' was 3.2 out of 5, which is below the average for similar introductory modules. Consider reviewing the way key programming concepts are presented. Make use of visual aids, analogies, and step-by-step explanations to enhance clarity.\"\n* **Recommendation 2 (Qualitative):** \"Multiple students commented that the concepts were confusing and the examples were too abstract.  Consider incorporating more real-world examples that are directly relatable to student's experiences. For instance, instead of using abstract variables, try a grocery shopping analogy.\"\n* **Recommendation 3 (Qualitative):** \"Students also suggested breaking down the code examples into smaller steps. For the next iteration, try creating code walkthroughs where you explain what each line does and why.\"\n* **Positive Reinforcement:** \"Students consistently praised your enthusiasm and passion for the subject.\"\n* **Next Steps:** \"We recommend reviewing the module materials and incorporating these suggestions for the next iteration. A follow-up session in a few weeks can further help you address any concerns.\"\n\n**Key Features of a Good Recommendation System:**\n\n* **User-Friendly Interface:** The tool should be easy for teachers to use and understand, even if they are not data experts.\n* **Customizable:** Allow teachers to focus on specific aspects of their teaching.\n* **Regular Updates:** The tool should provide updated data on a regular basis, allowing teachers to track their progress.\n* **Support and Resources:** Provide resources and training to help teachers understand the recommendations and implement changes.\n\n**Technology to Use:**\n\n* **Survey Platforms:** (e.g., Google Forms, SurveyMonkey, Typeform)\n* **Data Analysis Tools:** (e.g., Excel, Google Sheets, Python with Pandas, R)\n* **Visualization Tools:** (e.g., Tableau, Power BI)\n* **Natural Language Processing (NLP):**  For more advanced qualitative analysis\n\nBy implementing a comprehensive system for collecting, analyzing, and generating recommendations based on student feedback, teachers can gain valuable insights into the effectiveness of their modules, leading to a more engaging and effective learning experience for students. Remember that this is an iterative process. Regularly collect feedback and refine your approach to continually improve your teaching.\n",
          ],
        },
      ]
    )

     # Get the teacher feedback prompt
    teacher_prompt = compute_teacher_feedback_prompt(df, module_name)

    # Send the generated prompt to the model and get the response
    response = chat_session.send_message(f'''{teacher_prompt}''')

    # Return the response from the model
    return response.text
```
