```python
from transformers import pipeline

# Example DataFrame
import pandas as pd
data = {
    'review': ["Great course material!", "I loved the assignments and feedback", "The course structure is very clear"],
    'topic_summary': [['Course Material', 'Assignments'],
                      ['Assignments', 'Grading'],
                      ['Course Structure']]
}
df_test = pd.DataFrame(data)

# List of subtopics for each main topic (example, this would be specific to your dataset)
subtopics_dict = {
    'Course Material': ['Content Depth', 'Clarity', 'Engagement'],
    'Assignments': ['Difficulty', 'Variety', 'Workload'],
    'Grading': ['Fairness', 'Speed', 'Accuracy'],
    'Course Structure': ['Pacing', 'Organization', 'Flexibility']
}

# Load BART (or any model suitable for your subtopic classification task)
# For simplicity, we are using a sentiment-analysis pipeline, but you could train a custom model or fine-tune BART for subtopic classification.
# Load a zero-shot classifier from Hugging Face
subtopic_classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')

# Function to assign subtopics to each topic in a review
def assign_subtopics(row):
    result = {}  # Initialize an empty dictionary to store main topic and subtopic pairings

    for topic in row['topic_summary']:
        # Get the list of subtopics for the current main topic
        subtopics = subtopics_dict.get(topic, [])

        if subtopics:  # If there are subtopics to assign
            # Use BART to classify the subtopic (you could modify this logic as per your use case)
            # Using zero-shot classification to assign subtopic
            prediction = subtopic_classifier(row['review'], candidate_labels=subtopics)

            # Get the most likely subtopic based on the prediction
            best_subtopic = prediction['labels'][0]

            # Assign the subtopic to the main topic in the dictionary
            result[topic] = best_subtopic

    return result

# Apply the function to each row and create a new column with the result
df_test['topic_subtopic_dict'] = df_test.apply(assign_subtopics, axis=1)

# Display the resulting DataFrame
df_test
```

```python
from transformers import pipeline
import pandas as pd
import ast
from itertools import chain
import os
import pickle

# Load the zero-shot classifier
subtopic_classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')

# Ensure the DataFrame topic_summary is in list format
df = df_reviews.iloc[:500].copy()
# Ensure that topic_summary values are actual lists
df['topic_summary'] = df['topic_summary'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

unique_topics = list(set(chain.from_iterable(df['topic_summary'])))

subtopics_dict = {
    'Workload': ['difficulty', 'pace', 'manageability', 'accessibility', 'design'],
    'Course Engagement and Course Feelings(Sentiment)': [],
    'Utility and Usefulness': ['course structure', 'difficulty', 'support', 'teachers', 'clarity', 'key learnings'],
    'Course Material': ['difficulty', 'stimulation', 'usefulness', 'understanding', 'depth', 'scope', 'clarity', 'access'],
    'Instructor': ['teaching style', 'interaction', 'approachability', 'teaching clarity and accessibility'],
    'Lectures': ['content', 'organisation', 'usefulness', 'accessibility', 'structure', 'pace'],
    'Course Structure': [],
    'Exams': ['difficulty', 'content', 'time constraint', 'fairness'],
    'Grading': ['fairness', 'harshness'],
    'Assignments': ['difficulty', 'content', 'time constraint', 'fairness'],
    'Learning Outcomes': [],
    'Course Material and Understanding': ['difficulty', 'stimulation', 'usefulness', 'understanding', 'depth', 'scope', 'clarity', 'access']
}

def classify_row_batch(batch):
    """
    Classifies subtopics for each row in a batch of rows.
    """
    batch_results = []
    for _, row in batch.iterrows():
        result = {}
        for topic in row['topic_summary']:
            subtopics = subtopics_dict.get(topic, [])
            if subtopics:
                # Classify and assign the best subtopic
                prediction = subtopic_classifier(row['reviews'], candidate_labels=subtopics)
                result[topic] = prediction['labels'][0] if prediction['labels'] else None
        batch_results.append(result)
    return batch_results

def save_checkpoint(df, filename):
    """Save the checkpoint DataFrame to a pickle file."""
    with open(filename, 'wb') as f:
        pickle.dump(df, f)
    print(f"Checkpoint saved to {filename}")

def load_checkpoint(filename):
    """Load the checkpoint DataFrame from a pickle file."""
    with open(filename, 'rb') as f:
        return pickle.load(f)

# Check if checkpoint file exists
checkpoint_file = "checkpoint_labeled_reviews.pkl"
if os.path.exists(checkpoint_file):
    print("Loading from checkpoint...")
    df = load_checkpoint(checkpoint_file)
else:
    print("No checkpoint found, starting fresh...")

batch_size = 300
for start_idx in range(0, len(df), batch_size):
    end_idx = min(start_idx + batch_size, len(df))
    batch_df = df.iloc[start_idx:end_idx]

    print(f"Processing rows {start_idx} to {end_idx - 1}...")

    # Assign subtopics for this batch
    batch_results = classify_row_batch(batch_df)

    # Save results back to the DataFrame
    df.loc[start_idx:end_idx - 1, 'topic_subtopic_dict'] = batch_results

    # Save checkpoint after processing each batch
    save_checkpoint(df, checkpoint_file)

print("Processing complete.")
df[['reviews', 'topic_summary', 'topic_subtopic_dict']]
```

```python
from transformers import pipeline
import pandas as pd
import ast
from itertools import chain
import os
import torch

from google.colab import drive

# Mounting Google Drive
drive.mount('/content/drive', force_remount=True) # Adding force_remount=True to force the remounting process
import pandas as pd

# Load the zero-shot classifier
subtopic_classifier = pipeline('zero-shot-classification', model='facebook/bart-large-mnli')

df_reviews = pd.read_csv('/content/drive/MyDrive/merged_df_improved_labels.csv')

# Ensure the DataFrame topic_summary is in list format
df = df_reviews.copy()
# Ensure that topic_summary values are actual lists
df['topic_summary'] = df['topic_summary'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

unique_topics = list(set(chain.from_iterable(df['topic_summary'])))

subtopics_dict = {
    'Workload': ['difficulty', 'pace', 'manageability', 'accessibility', 'design'],
    'Course Engagement and Course Feelings(Sentiment)': [],
    'Utility and Usefulness': ['course structure', 'difficulty', 'support', 'teachers', 'clarity', 'key learnings'],
    'Course Material': ['difficulty', 'stimulation', 'usefulness', 'understanding', 'depth', 'scope', 'clarity', 'access'],
    'Instructor': ['teaching style', 'interaction', 'approachability', 'teaching clarity and accessibility'],
    'Lectures': ['content', 'organisation', 'usefulness', 'accessibility', 'structure', 'pace'],
    'Course Structure': [],
    'Exams': ['difficulty', 'content', 'time constraint', 'fairness'],
    'Grading': ['fairness', 'harshness'],
    'Assignments': ['difficulty', 'content', 'time constraint', 'fairness'],
    'Learning Outcomes': [],
    'Course Material and Understanding': ['difficulty', 'stimulation', 'usefulness', 'understanding', 'depth', 'scope', 'clarity', 'access']
}

def classify_row_batch(batch):
    """
    Classifies subtopics for each row in a batch of rows.
    """
    batch_results = []
    for _, row in batch.iterrows():
        result = {}
        for topic in row['topic_summary']:
            subtopics = subtopics_dict.get(topic, [])
            if subtopics:
                # Classify and assign the best subtopic
                prediction = subtopic_classifier(row['reviews'], candidate_labels=subtopics)
                result[topic] = prediction['labels'][0] if prediction['labels'] else None
        batch_results.append(result)
    return batch_results

def save_checkpoint(df_batch, filename):
    """Save the checkpoint DataFrame batch to a .pth file."""
    checkpoint = {"data": df_batch}
    torch.save(checkpoint, filename)
    print(f"Checkpoint saved to {filename}")

def load_checkpoint(filename):
    """Load the checkpoint DataFrame batch from a .pth file."""
    checkpoint = torch.load(filename)
    return checkpoint["data"]

# Directory for checkpoints
checkpoint_dir = "/content/drive/MyDrive/checkpoints_tcs/"
os.makedirs(checkpoint_dir, exist_ok=True)

batch_size = 500

for batch_num, start_idx in enumerate(range(0, len(df), batch_size)):
    end_idx = min(start_idx + batch_size, len(df))
    batch_df = df.iloc[start_idx:end_idx]

    checkpoint_file = os.path.join(checkpoint_dir, f"checkpoint_batch_{batch_num + 1}.pth")

    # Skip processing if the checkpoint already exists
    if os.path.exists(checkpoint_file):
        print(f"Checkpoint already exists for batch {batch_num + 1}, skipping...")
        continue

    print(f"Processing rows {start_idx} to {end_idx - 1}...")

    # Assign subtopics for this batch
    batch_results = classify_row_batch(batch_df)

    # Save results back to the batch DataFrame
    batch_df['topic_subtopic_dict'] = batch_results

    # Save checkpoint for this batch
    save_checkpoint(batch_df, checkpoint_file)

print("Processing complete.")
```

```python
import os
import torch

# Directory where your checkpoint files are stored
checkpoint_dir = "/content/drive/MyDrive/checkpoints_tcs/"

# Initialize an empty list to hold the batch data
all_batch_data = []

# Iterate over all the checkpoint files
for checkpoint_file in os.listdir(checkpoint_dir):
    if checkpoint_file.endswith('.pth'):
        checkpoint_path = os.path.join(checkpoint_dir, checkpoint_file)

        # Load the checkpoint
        checkpoint = torch.load(checkpoint_path)
        batch_df = checkpoint["data"]

        # Append the batch data (with 'topic_subtopic_dict') to the list
        all_batch_data.append(batch_df)

# Concatenate all batches to form the full DataFrame
df_full = pd.concat(all_batch_data, ignore_index=True)

# Now df_full contains all the rows from your original DataFrame along with 'topic_subtopic_dict'

# If you want to update the original df (df_reviews), you can do the following:
df['topic_subtopic_dict'] = df_full['topic_subtopic_dict']

# Now, df_reviews should have the 'topic_subtopic_dict' column populated with the results from the checkpoints.
```
