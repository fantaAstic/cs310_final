```python
import ast

# Function to flatten and combine topics from strings in both columns
def flatten_and_combine(row):
    topics = []

    for col in ["topic_summary", "topic_summary_v2"]:
        if isinstance(row[col], str):  # Check if the entry is a string
            try:
                # Convert the string representation of the list to an actual list
                topic_list = ast.literal_eval(row[col])
                # Ensure it is a list and extend the topics list
                if isinstance(topic_list, list):
                    topics.extend(topic_list)
            except (ValueError, SyntaxError):
                # If conversion fails, handle the error (perhaps skip or log)
                print(f"Error parsing string into list for row with index {row.name}")

    return list(set(topics))  # Remove duplicates and return as a list

# Apply the function to create the "all_topics" column
full_df["all_topics"] = full_df.apply(flatten_and_combine, axis=1)

# Display the result
print(full_df[["all_topics"]].head())
```

```python
from collections import Counter
import pandas as pd
from collections import Counter
import ast
import os
import google.generativeai as genai
import matplotlib.pyplot as plt

def compute_module_id(module_name: str) -> int: # should return a number, can be a random number tbh
    """Generate a unique ID for a module based on its name."""
    pass

def compute_summary(df: pd.DataFrame, module_name: str) -> str:
    """Generate a summary of reviews, sentiment trends, and emotions for a given module."""
    # Step 1: Group by 'month' and calculate the average emotion score for each month
    curr_module_reviews = df[df['name'] == module_name].copy()  # Use .copy() to avoid warning

    # Assuming 'emotion_label' contains categorical labels (not scores), we might need to adjust the aggregation
    # Here we can count the occurrences of each emotion per month
    monthly_emotion_counts = curr_module_reviews.groupby(['month', 'emotion_label']).size().unstack(fill_value=0)
    monthly_emotion_trend = monthly_emotion_counts.diff(axis=0).max(axis=1)  # This finds if there's any increase in emotion counts

    # Find the last month where emotion count improved
    last_emotion_improvement_month = monthly_emotion_trend[monthly_emotion_trend > 0].idxmax()

    # If it's a string, convert it to datetime
    if isinstance(last_emotion_improvement_month, str):
        last_emotion_improvement_month = pd.to_datetime(last_emotion_improvement_month, format='%Y-%m')

    # Step 2: Group by 'month' and calculate the average rating for each month
    monthly_avg_rating = curr_module_reviews.groupby('month')['rating'].mean()
    monthly_avg_rating_diff = monthly_avg_rating.diff()

    # Find the last month where rating improved
    last_rating_improvement_month = monthly_avg_rating_diff[monthly_avg_rating_diff > 0].idxmax()

    # If it's a string, convert it to datetime
    if isinstance(last_rating_improvement_month, str):
        last_rating_improvement_month = pd.to_datetime(last_rating_improvement_month, format='%Y-%m')

    # Step 3: Flatten the lists from the 'topic_summary' column into one long list
    all_topics = [topic for sublist in curr_module_reviews['all_topics'] for topic in sublist]
    topic_counts = Counter(all_topics)
    most_common_topic = topic_counts.most_common(1)[0]

    # Step 4: Get the most recent month
    curr_module_reviews.loc[:, 'month'] = pd.to_datetime(curr_module_reviews['month'], format='%Y-%m')  # Use .loc to modify safely
    most_recent_month = curr_module_reviews['month'].max()

    # Step 5: Find the most popular sentiment and emotion for the most recent month
    recent_month_reviews = curr_module_reviews[curr_module_reviews['month'] == most_recent_month]
    most_recent_sentiment = recent_month_reviews['sentiment_label_from_review'].value_counts().idxmax()
    most_recent_emotion = recent_month_reviews['emotion_label'].value_counts().idxmax()

    # Step 6: Count consecutive months where the sentiment and emotion stayed the same
    same_rating_months = count_consecutive_months(curr_module_reviews, most_recent_month, 'rating')
    same_emotion_months = count_consecutive_months(curr_module_reviews, most_recent_month, 'emotion_label')

    # Prepare the final outlook string
    summary_str = f'''
    Most reviews are {most_recent_sentiment} and about {most_common_topic[0]} with most students feeling {most_recent_emotion}.
    There has been no improvement in sentiment since {last_rating_improvement_month.strftime('%Y-%m')} and the last improvement in emotion
    was in {last_emotion_improvement_month.strftime('%Y-%m')}.
    The most recent reviews are {most_recent_sentiment} and {most_recent_emotion}.
    Over the existence of the course, the number of {most_recent_sentiment} and {most_recent_emotion} reviews has increased.
    The sentiment has stayed the same for {same_rating_months} consecutive month(s) and the emotion has stayed the same for
    {same_emotion_months} consecutive month(s).
    '''

    # Return or print the final summary string
    return summary_str



def count_consecutive_months(df, most_recent_month, column):
    # Filter data from the most recent month going backward
    df_sorted = df[df['month'] <= most_recent_month].sort_values('month', ascending=False)

    # Get the most recent value for the column
    recent_value = df_sorted[df_sorted['month'] == most_recent_month][column].values[0]

    # Count consecutive months where the value stays the same
    count = 0
    for value in df_sorted[column]:
        if value == recent_value:
            count += 1
        else:
            break
    return count

def compute_outlook(df: pd.DataFrame, module_name: str) -> str:
    """Determine the outlook (e.g., 'positive', 'negative', 'mixed') for a module based on sentiment trends."""

    # Filter the reviews for the specified module
    module_reviews = df[df['name'] == module_name]

    # Calculate the percentage of positive reviews (assuming 'positive' is the label for positive sentiment)
    positive_reviews = module_reviews[module_reviews['sentiment_label_from_review'] == 'positive']
    total_reviews = len(module_reviews)

    # Calculate percentage
    positive_percentage = (len(positive_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    # Determine the outlook based on the percentage of positive reviews
    if positive_percentage > 70:
        outlook = 'Positive'
    elif positive_percentage < 50:
        outlook = 'Negative'
    else:
        outlook = 'Mixed'

    return f"The outlook for '{module_name}' is: {outlook}"


def compute_positive_reviews(df: pd.DataFrame, module_name: str) -> str:
    """Calculate the percentage of positive reviews for a module."""
    # Filter the reviews for the specified module
    module_reviews = df[df['name'] == module_name]

    # Count positive reviews (Assume 'positive' is the label for positive reviews)
    positive_reviews = module_reviews[module_reviews['sentiment_label_from_review'] == 'positive']
    total_reviews = len(module_reviews)

    # Calculate percentage
    positive_percentage = (len(positive_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    # Return the result as a formatted string
    return f"Percentage of positive reviews for '{module_name}': {positive_percentage:.2f}%"

def compute_negative_reviews(df: pd.DataFrame, module_name: str) -> str:
    """Calculate the percentage of negative reviews for a module."""
    # Filter the reviews for the specified module
    module_reviews = df[df['name'] == module_name]

    # Count negative reviews (Assume 'negative' is the label for negative reviews)
    negative_reviews = module_reviews[module_reviews['sentiment_label_from_review'] == 'negative']
    total_reviews = len(module_reviews)

    # Calculate percentage
    negative_percentage = (len(negative_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    # Return the result as a formatted string
    return f"Percentage of negative reviews for '{module_name}': {negative_percentage:.2f}%"

def get_category(df: pd.DataFrame, module_name: str) -> str:
    """Retrieve the category of a module based on its name."""

    # Find the row where the 'name' column matches the provided module_name
    row = df[df['name'].str.lower() == module_name.lower()]

    # If a match is found, return the category
    if not row.empty:
        return row['category'].iloc[0]
    else:
        # Return a default category if no match is found
        return "Uncategorized"

def compute_teacher_feedback_prompt(df: pd.DataFrame, module_name: str) -> str:
    """Generate a feedback prompt for a given module based on the feedback summary in the dataset."""

    # Filter the data for the specific module
    module_reviews = df[df['name'] == module_name]

    # Assuming 'rating' represents overall feedback and sentiment column is 'sentiment_label_from_review'
    # Adjust these as per your actual dataset columns
    avg_rating = module_reviews['rating'].mean()
    sentiment_percentage = (module_reviews['sentiment_label_from_review'] == 'positive').mean() * 100

    # Get the most common emotions (you can adjust how to retrieve emotions)
    # For example, assuming 'emotion_label' contains emotion data
    most_common_emotions = module_reviews['emotion_label'].mode().head(3).tolist()

    # Get the main areas of commentary from 'topic_subtopic_dict' column
    # Ensure it's properly converted from string to dictionary
    subtopics = ast.literal_eval(module_reviews['topic_subtopic_dict'].iloc[0])  # Convert string to dictionary

    # Format the commentary for each subtopic
    formatted_commentary = "\n".join([f"{key}: {value}" for key, value in subtopics.items()])

    # For time trend, assuming you have 'month' column (if not, adjust based on your dataset)
    # You could calculate sentiment or rating trends over months, for simplicity, let's assume positive trend:
    # Example logic for time trend improvement (adjust based on actual data)
    time_trend = "Positive feedback consistently increased throughout the semester, especially after midterm."

    outlook = compute_outlook(df, module_name)

    # Construct the prompt
     f'''
    I’m a teacher of a '{module_name}' module, and I recently received feedback with the following summary:

    Overall: {outlook} feedback ({avg_rating:.1f} average rating, {sentiment_percentage:.1f}% positive sentiment).

    Key (most common) Emotions: {', '.join(most_common_emotions)}.
prompt =
    Main Areas of Commentary:
    {formatted_commentary}

    Time Trend: {time_trend}

    I need guidance on how to build on this positive feedback and further enhance the student experience. Specifically, I'd like recommendations that are:

    Specific & Actionable: For each area of positive commentary (e.g., Exam, Assignment, Lecturer, and Grading), provide me with practical,
    specific actions I can take to maintain or improve these areas.
    Preventative: How can I prevent complacency and continue improving these areas in future semesters or iterations of the module?
    Emotionally Sensitive: How can I acknowledge the positive feedback and use it to motivate students further, while maintaining a
    positive learning environment?
    Prioritised: If there are multiple things I could do to improve, which areas should I focus on first and why?
    Considered for Time Trend: How can I use the positive time trend to ensure that future semesters continue on this upward trajectory?

    I’m looking for a response that is practical, forward-thinking, and specific to my situation as a teacher for a '{module_name}' module.
    '''

    return prompt

def compute_teacher_feedback_recommendation(df: pd.DataFrame, module_name: str) -> str:
    """Generate recommendations for teachers based on the feedback analysis for a module."""
    genai.configure(api_key='SECRET')

    # Create the model
    generation_config = {
      "temperature": 0.7, # need to check temp param
      "top_p": 0.95,
      "top_k": 40,
      "max_output_tokens": 8192,
      "response_mime_type": "text/plain",
    }

    model = genai.GenerativeModel(
      model_name="gemini-2.0-flash-exp",
      generation_config=generation_config,
    )

    chat_session = model.start_chat(
      history=[
        {
          "role": "user",
          "parts": [
            "something to give recommendations to teachers to improve modules based on student feedback on modules that they teach\n",
          ],
        },
        {
          "role": "model",
          "parts": [
            "Okay, let's craft something that can provide teachers with actionable recommendations based on student feedback on their modules. Here's a breakdown of how it could work, along with examples:\n\n**Core Components:**\n\n1. **Feedback Collection:**\n   * **Structured Surveys:** This is the most common method. Questions could cover:\n      * **Clarity of Content:** (e.g., Was the material presented in a clear and understandable way?)\n      * **Engagement:** (e.g., Did you find the module engaging? Did it hold your interest?)\n      * **Organization:** (e.g., Was the module well-organized? Did the flow of topics make sense?)\n      * **Usefulness:** (e.g., How relevant was the material to your learning goals?)\n      * **Pacing:** (e.g., Was the pace of the module appropriate?)\n      * **Assessment:** (e.g., Were the assessments fair and aligned with the content?)\n      * **Instructor's Role:** (e.g., Was the instructor responsive and helpful?)\n      * **Overall Experience:** (e.g., How satisfied were you with the module overall?)\n   * **Open-Ended Feedback:** Allow students to provide written comments on what they liked, what they didn't, and suggestions for improvement.\n   * **Anonymous Feedback:** Encourage honest responses by ensuring anonymity.\n\n2. **Data Analysis:**\n   * **Quantitative Analysis:**\n      * **Average Scores:** Calculate average scores for each question to identify areas of strength and weakness.\n      * **Frequency Distributions:** Analyze how often students selected specific responses (e.g., how many students rated a question as \"Excellent\").\n      * **Trend Analysis:** Track changes in scores over time to monitor improvement or decline in different areas.\n   * **Qualitative Analysis:**\n      * **Theme Identification:** Look for common themes and patterns in the open-ended feedback.\n      * **Sentiment Analysis:** Use tools to analyze the overall tone (positive, negative, neutral) of the comments.\n\n3. **Recommendation Generation:** This is the key part. Here's how to generate meaningful recommendations:\n   * **Based on Quantitative Data:**\n      * **Low-Scoring Areas:** If a question consistently scores low, suggest specific actions (e.g., \"The question 'Clarity of Content' scored below average. Consider providing more detailed examples or explanations.\")\n      * **Inconsistent Ratings:** If there's a wide spread of ratings, this might indicate that the content appeals to some but not others. Recommend offering optional extensions or different approaches to the material.\n   * **Based on Qualitative Data:**\n       * **Recurring Themes:** If multiple students mention that the pacing is too fast, recommend slowing down. If multiple students mention that the instructor's responsiveness is good, highlight this strength.\n       * **Specific Suggestions:** If a student suggests adding more real-world examples, suggest incorporating them.\n   * **Focus on Actionable Advice:** Recommendations should be specific and easily translated into concrete changes. Avoid vague statements.\n   * **Provide Context:** Present data in a way that is clear and understandable for the teacher, along with comparisons to previous years or other similar modules (if available).\n   * **Highlight Strengths:**  Don't focus solely on weaknesses.  Also include positive feedback to reinforce what is working well.\n\n**Example Scenario and Recommendations:**\n\n**Scenario:**\n\n* **Module:** \"Introduction to Programming\"\n* **Feedback:**\n   * **Quantitative:**  Low average score on the \"Clarity of Content\" question.\n   * **Qualitative:**  Multiple students commented that \"the concepts were confusing\" and \"the examples were too abstract.\"\n\n**Generated Recommendations:**\n\n* **Recommendation 1 (Quantitative):**  \"The average score for 'Clarity of Content' was 3.2 out of 5, which is below the average for similar introductory modules. Consider reviewing the way key programming concepts are presented. Make use of visual aids, analogies, and step-by-step explanations to enhance clarity.\"\n* **Recommendation 2 (Qualitative):** \"Multiple students commented that the concepts were confusing and the examples were too abstract.  Consider incorporating more real-world examples that are directly relatable to student's experiences. For instance, instead of using abstract variables, try a grocery shopping analogy.\"\n* **Recommendation 3 (Qualitative):** \"Students also suggested breaking down the code examples into smaller steps. For the next iteration, try creating code walkthroughs where you explain what each line does and why.\"\n* **Positive Reinforcement:** \"Students consistently praised your enthusiasm and passion for the subject.\"\n* **Next Steps:** \"We recommend reviewing the module materials and incorporating these suggestions for the next iteration. A follow-up session in a few weeks can further help you address any concerns.\"\n\n**Key Features of a Good Recommendation System:**\n\n* **User-Friendly Interface:** The tool should be easy for teachers to use and understand, even if they are not data experts.\n* **Customizable:** Allow teachers to focus on specific aspects of their teaching.\n* **Regular Updates:** The tool should provide updated data on a regular basis, allowing teachers to track their progress.\n* **Support and Resources:** Provide resources and training to help teachers understand the recommendations and implement changes.\n\n**Technology to Use:**\n\n* **Survey Platforms:** (e.g., Google Forms, SurveyMonkey, Typeform)\n* **Data Analysis Tools:** (e.g., Excel, Google Sheets, Python with Pandas, R)\n* **Visualization Tools:** (e.g., Tableau, Power BI)\n* **Natural Language Processing (NLP):**  For more advanced qualitative analysis\n\nBy implementing a comprehensive system for collecting, analyzing, and generating recommendations based on student feedback, teachers can gain valuable insights into the effectiveness of their modules, leading to a more engaging and effective learning experience for students. Remember that this is an iterative process. Regularly collect feedback and refine your approach to continually improve your teaching.\n",
          ],
        },
      ]
    )

     # Get the teacher feedback prompt
    teacher_prompt = compute_teacher_feedback_prompt(df, module_name)

    # Send the generated prompt to the model and get the response
    response = chat_session.send_message(f'''{teacher_prompt}''')

    # Return the response from the model
    return response.text

import os
import matplotlib.pyplot as plt
import pandas as pd

# Define a dictionary for emotion to number mapping
emotion_to_number = {
    'joy': 10,
    'love': 9,
    'enjoyment/happiness': 8,
    'amusement': 7,
    'excitement': 7,
    'optimism': 6,
    'pride': 6,
    'gratitude': 5,
    'caring': 5,
    'relief': 4,
    'approval': 4,
    'desire': 3,
    'curiosity': 2,
    'surprise': 1,
    'realization': 1,
    'confusion': -1,
    'nervousness': -2,
    'annoyance': -3,
    'disapproval': -4,
    'disappointment': -5,
    'sadness': -6,
    'remorse': -6,
    'embarrassment': -7,
    'disgust': -8,
    'anger': -9,
    'fear': -10
}

import re

def plot_time_series_sentiment(full_df_need: pd.DataFrame, output_dir: str, module_name: str) -> str:
    """Generate and show a plot of the most common sentiment per month for a given topic."""
    # Group by month and sentiment, then count the occurrences of each sentiment
    monthly_emotion_counts = full_df_need.groupby(['month', 'sentiment_label_from_review']).size().unstack(fill_value=0)

    # Find the most common sentiment for each month
    most_common_sentiment_per_month = monthly_emotion_counts.idxmax(axis=1)

    # Find the count of the most common sentiment for each month
    most_common_sentiment_count_per_month = monthly_emotion_counts.max(axis=1)

    # Handle NaN values by filling them with a placeholder if needed
    most_common_sentiment_per_month = most_common_sentiment_per_month.fillna('Unknown')

    # Create a plot for the most common sentiment per month
    plt.figure(figsize=(30, 14))  # Adjusted figure size to fit most screen sizes

    # Create a bar plot
    bars = plt.bar(most_common_sentiment_per_month.index, most_common_sentiment_count_per_month, width=0.6)

    # Add titles and labels dynamically based on module_name
    plt.title(f'Most Common Sentiment for {module_name} Courses (Monthly)', fontsize=14)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Number of Reviews', fontsize=12)

    # Rotate x-axis labels to prevent overlap
    plt.xticks(rotation=45, ha='right', fontsize=10)

    # Add labels to each bar indicating the sentiment
    for bar, sentiment in zip(bars, most_common_sentiment_per_month):
        y_position = bar.get_height() + 10 if bar.get_height() > 10 else bar.get_height() - 5
        plt.text(bar.get_x() + bar.get_width() / 2, y_position, sentiment,
                 ha='center', va='bottom' if bar.get_height() > 10 else 'top', fontsize=10, rotation=90)

    # Show the plot before saving
    plt.tight_layout()

    # Clean the module name for the filename: replace invalid characters with dashes
    safe_module_name = re.sub(r'[^\w\s-]', '', module_name)  # Remove any special characters
    safe_module_name = re.sub(r'[-\s]+', '-', safe_module_name)  # Replace spaces and multiple dashes with a single dash

    # Create the filename using the cleaned module name
    plot_filename = f'{output_dir}sentiment_timeseries_{safe_module_name}.png'

    # Debug statement to print the file path
    print(f"Saving plot to {plot_filename}")

    # Save the plot to a file after rendering
    plt.savefig(plot_filename, dpi=300, bbox_inches='tight', pad_inches=0.1)

    # Display the plot
    plt.show()

    # Close the figure after saving
    plt.close()

    return plot_filename

def plot_time_series_emotion(full_df_need: pd.DataFrame, output_dir: str, module_name: str) -> str:
    """Generate and show a plot of the most common emotion per month for a given topic."""

    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Group by month and emotion, then count the occurrences of each emotion
    monthly_emotion_counts = full_df_need.groupby(['month', 'emotion_label']).size().unstack(fill_value=0)

    # Find the most common emotion for each month
    most_common_emotion_per_month = monthly_emotion_counts.idxmax(axis=1)

    # Find the count of the most common emotion for each month
    most_common_emotion_count_per_month = monthly_emotion_counts.max(axis=1)

    # Map emotions to their respective numbers for better visualization
    most_common_emotion_numeric = most_common_emotion_per_month.map(emotion_to_number)

    # Create a plot for the most common emotion per month
    plt.figure(figsize=(30, 14))  # Adjusted figure size for better visibility

    # Create a bar plot
    bars = plt.bar(most_common_emotion_per_month.index, most_common_emotion_count_per_month, color='skyblue', width=0.6)

    # Add titles and labels
    plt.title(f'Most Common Emotion for {module_name} (Monthly)', fontsize=14)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Number of Reviews', fontsize=12)

    # Rotate x-axis labels to prevent overlap
    plt.xticks(rotation=45, ha='right', fontsize=10)

    # Add labels to each bar indicating the emotion number (mapped value)
    for bar, emotion, numeric_emotion in zip(bars, most_common_emotion_per_month, most_common_emotion_numeric):
        y_position = bar.get_height() + 10 if bar.get_height() > 10 else bar.get_height() - 5
        plt.text(bar.get_x() + bar.get_width() / 2, y_position, f'{emotion} ({numeric_emotion})',
                 ha='center', va='bottom' if bar.get_height() > 10 else 'top', fontsize=10, rotation=90)

    # Show the plot before saving
    plt.tight_layout()

    # Clean the module name for the filename: replace invalid characters with dashes
    safe_module_name = re.sub(r'[^\w\s-]', '', module_name)  # Remove any special characters
    safe_module_name = re.sub(r'[-\s]+', '-', safe_module_name)  # Replace spaces and multiple dashes with a single dash

    # Create the filename using the cleaned module name
    plot_filename = f'{output_dir}emotion_timeseries_{safe_module_name}.png'

    # Debug statement to print the file path
    print(f"Saving plot to {plot_filename}")

    # Save the plot to a file after rendering
    plt.savefig(plot_filename, dpi=300, bbox_inches='tight', pad_inches=0.1)

    # Display the plot
    plt.show()

    # Close the figure after saving
    plt.close()

    return plot_filename

def compute_analysis_refs(module_name: str, full_df_need: pd.DataFrame, output_dir: str) -> str:
    """Generate analysis reference names for sentiment and emotion time series data for a given topic."""
    # Assuming the module_name is relevant for the time series data plots, call the plot functions
    emotion_plot_file = plot_time_series_emotion(full_df_need, output_dir, module_name)
    sentiment_plot_file = plot_time_series_sentiment(full_df_need, output_dir, module_name)

    # Create a tuple of the filenames (or analysis references)
    analysis_refs = (emotion_plot_file, sentiment_plot_file)

    # Return the file names as a string (or can be returned as a tuple if needed)
    return analysis_refs

def compute_positive_emotion_reviews(df: pd.DataFrame, module_name: str) -> str:
    """Calculate the percentage of reviews with positive emotions for a module."""
    # Filter reviews for the given module
    module_reviews = df[df['name'] == module_name]

    # Count reviews with positive emotions (emotion score >= 4)
    positive_reviews = module_reviews[module_reviews['emotion_label'].map(emotion_to_number) >= 4]
    total_reviews = len(module_reviews)

    # Compute percentage
    positive_percentage = (len(positive_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    return f"Percentage of reviews with positive emotions for '{module_name}': {positive_percentage:.2f}%"


def compute_negative_emotion_reviews(df: pd.DataFrame, module_name: str) -> str:
    """Calculate the percentage of reviews with negative emotions for a module."""
    # Filter reviews for the given module
    module_reviews = df[df['name'] == module_name]

    # Count reviews with negative emotions (emotion score < 4)
    negative_reviews = module_reviews[module_reviews['emotion_label'].map(emotion_to_number) < 4]
    total_reviews = len(module_reviews)

    # Compute percentage
    negative_percentage = (len(negative_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    return f"Percentage of reviews with negative emotions for '{module_name}': {negative_percentage:.2f}%"

```

```python
import pandas as pd
from collections import Counter
from typing import List
import ast
import re

def compute_topics(df: pd.DataFrame, module_name: str) -> List[str]:
    """Extract relevant topics from the module's reviews."""
    module_df = df[df['name'] == module_name]

    # Flatten topics (assuming topic_summary contains lists or comma-separated strings)
    all_topics = set()
    for topics in module_df['topic_summary'].dropna():
        if isinstance(topics, list):
            all_topics.update(topics)
        else:  # Assume comma-separated string or string representation of a list
            try:
                # Try to convert string representation of a list into an actual list
                topics_list = ast.literal_eval(topics)
                if isinstance(topics_list, list):  # Ensure it's a list after conversion
                    all_topics.update(topics_list)
            except (ValueError, SyntaxError):
                # If there's an error, split by commas and clean up
                all_topics.update(map(str.strip, topics.split(',')))

    return list(all_topics)



def compute_topic_summary(df: pd.DataFrame, module_name: str, topic: str) -> str:
    """Generate a summary of sentiment trends and emotions for a specific topic within a module."""
    # Filter reviews for the given module and topic
    topic_reviews = df[(df['name'] == module_name) & (df['topic_summary'].apply(lambda x: topic in x))].copy()

    if topic_reviews.empty:
        return f"No reviews found for topic '{topic}' in module '{module_name}'."

    # Ensure 'month' is in datetime format
    topic_reviews['month'] = pd.to_datetime(topic_reviews['month'], format='%Y-%m', errors='coerce')

    # Handle any missing or invalid month entries
    if topic_reviews['month'].isna().any():
        return f"Some 'month' values couldn't be parsed as datetime for topic '{topic}' in module '{module_name}'."

    # Step 1: Most recent sentiment/emotion
    most_recent_month = topic_reviews['month'].max()
    recent_reviews = topic_reviews[topic_reviews['month'] == most_recent_month]
    most_recent_sentiment = recent_reviews['sentiment_label_from_review'].value_counts().idxmax()
    most_recent_emotion = recent_reviews['emotion_label'].value_counts().idxmax()

    # Step 2: Track sentiment and emotion trends over time
    sentiment_counts_by_month = topic_reviews.groupby(['month', 'sentiment_label_from_review']).size().unstack(fill_value=0)
    emotion_counts_by_month = topic_reviews.groupby(['month', 'emotion_label']).size().unstack(fill_value=0)

    sentiment_trend = sentiment_counts_by_month.diff(axis=0)
    emotion_trend = emotion_counts_by_month.diff(axis=0)

    # Step 3: Find last sentiment/emotion improvement months
    sentiment_improvement_month = sentiment_trend[sentiment_trend > 0].dropna().index.max() if not sentiment_trend.empty else pd.NaT
    emotion_improvement_month = emotion_trend[emotion_trend > 0].dropna().index.max() if not emotion_trend.empty else pd.NaT

    # Step 4: Consecutive months with same sentiment/emotion
    same_sentiment_months = count_consecutive_months(topic_reviews, most_recent_month, 'sentiment_label_from_review')
    same_emotion_months = count_consecutive_months(topic_reviews, most_recent_month, 'emotion_label')

    # Prepare the final summary string
    summary_str = f'''
    For the topic '{topic}' in '{module_name}':
    - Most recent sentiment: {most_recent_sentiment}
    - Most recent emotion: {most_recent_emotion}
    - Last sentiment improvement: {sentiment_improvement_month.strftime('%Y-%m') if pd.notna(sentiment_improvement_month) else 'N/A'}
    - Last emotion improvement: {emotion_improvement_month.strftime('%Y-%m') if pd.notna(emotion_improvement_month) else 'N/A'}
    - Sentiment has stayed the same for {same_sentiment_months} consecutive months.
    - Emotion has stayed the same for {same_emotion_months} consecutive months.
    '''

    return summary_str

def compute_topic_outlook(df: pd.DataFrame, module_name: str, topic: str) -> str:
    """Determine the outlook for a specific topic based on sentiment trends."""
    topic_reviews = df[(df['name'] == module_name) & (df['topic_summary'].apply(lambda x: topic in x))]

    if topic_reviews.empty:
        return f"No reviews found for topic '{topic}' in module '{module_name}'."

    positive_reviews = topic_reviews[topic_reviews['sentiment_label_from_review'] == 'positive']
    total_reviews = len(topic_reviews)

    positive_percentage = (len(positive_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    if positive_percentage > 70:
        outlook = 'Positive'
    elif positive_percentage < 50:
        outlook = 'Negative'
    else:
        outlook = 'Mixed'

    return f"The outlook for topic '{topic}' in '{module_name}' is: {outlook}"


def compute_topic_positive_reviews(df: pd.DataFrame, module_name: str, topic: str) -> str:
    """Calculate the percentage of positive reviews for a specific topic."""
    topic_reviews = df[(df['name'] == module_name) & (df['topic_summary'].apply(lambda x: topic in x))]

    if topic_reviews.empty:
        return f"No reviews found for topic '{topic}' in module '{module_name}'."

    positive_reviews = topic_reviews[topic_reviews['sentiment_label_from_review'] == 'positive']
    total_reviews = len(topic_reviews)

    positive_percentage = (len(positive_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    return f"Percentage of positive reviews for topic '{topic}' in '{module_name}': {positive_percentage:.2f}%"


def compute_topic_negative_reviews(df: pd.DataFrame, module_name: str, topic: str) -> str:
    """Calculate the percentage of negative reviews for a specific topic."""
    topic_reviews = df[(df['name'] == module_name) & (df['topic_summary'].apply(lambda x: topic in x))]

    if topic_reviews.empty:
        return f"No reviews found for topic '{topic}' in module '{module_name}'."

    negative_reviews = topic_reviews[topic_reviews['sentiment_label_from_review'] == 'negative']
    total_reviews = len(topic_reviews)

    negative_percentage = (len(negative_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    return f"Percentage of negative reviews for topic '{topic}' in '{module_name}': {negative_percentage:.2f}%"

def compute_positive_emotion_reviews_topic(df: pd.DataFrame, module_name: str, topic: str) -> str:
    """Calculate the percentage of reviews with positive emotions for a module."""
    # Filter reviews for the given module
    module_reviews = df[(df['name'] == module_name) & (df['topic_summary'].apply(lambda x: topic in x))]

    # Count reviews with positive emotions (emotion score >= 4)
    positive_reviews = module_reviews[module_reviews['emotion_label'].map(emotion_to_number) >= 4]
    total_reviews = len(module_reviews)

    # Compute percentage
    positive_percentage = (len(positive_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    return f"Percentage of reviews with positive emotions for '{topic} in {module_name}': {positive_percentage:.2f}%"


def compute_negative_emotion_reviews_topic(df: pd.DataFrame, module_name: str, topic: str) -> str:
    """Calculate the percentage of reviews with negative emotions for a module."""
    # Filter reviews for the given module
    module_reviews = df[(df['name'] == module_name) & (df['topic_summary'].apply(lambda x: topic in x))]

    # Count reviews with negative emotions (emotion score < 4)
    negative_reviews = module_reviews[module_reviews['emotion_label'].map(emotion_to_number) < 4]
    total_reviews = len(module_reviews)

    # Compute percentage
    negative_percentage = (len(negative_reviews) / total_reviews) * 100 if total_reviews > 0 else 0

    return f"Percentage of reviews with negative emotions for '{topic} in {module_name}': {negative_percentage:.2f}%"


def count_consecutive_months(df, most_recent_month, column):
    """Count consecutive months where the value in the given column stays the same."""
    df_sorted = df[df['month'] <= most_recent_month].sort_values('month', ascending=False)

    if df_sorted.empty:
        return 0

    recent_value = df_sorted[df_sorted['month'] == most_recent_month][column].values[0]

    count = 0
    for value in df_sorted[column]:
        if value == recent_value:
            count += 1
        else:
            break
    return count


import os

def plot_time_series_sentiment_topic(df: pd.DataFrame, output_dir: str, module_name: str, topic: str) -> str:
    """Generate and show a plot of the most common sentiment per month for a given topic."""
    # Group by month and sentiment, then count the occurrences of each sentiment
    df = df[
        (df['name'] == module_name) &  # Filter by module name
        (df['topic_summary'].astype(str).str.contains(topic, na=False))  # Filter by topic in topic_summary
    ]
    monthly_emotion_counts = df.groupby(['month', 'sentiment_label_from_review']).size().unstack(fill_value=0)

    # Find the most common sentiment for each month
    most_common_sentiment_per_month = monthly_emotion_counts.idxmax(axis=1)

    # Find the count of the most common sentiment for each month
    most_common_sentiment_count_per_month = monthly_emotion_counts.max(axis=1)

    # Handle NaN values by filling them with a placeholder if needed
    most_common_sentiment_per_month = most_common_sentiment_per_month.fillna('Unknown')

    # Create a plot for the most common sentiment per month
    plt.figure(figsize=(30, 14))  # Adjusted figure size to fit most screen sizes

    # Create a bar plot
    bars = plt.bar(most_common_sentiment_per_month.index, most_common_sentiment_count_per_month, width=0.6)

    # Add titles and labels dynamically based on module_name
    plt.title(f'Most Common Sentiment for {topic} ({module_name}) Courses (Monthly)', fontsize=14)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Number of Reviews', fontsize=12)

    # Rotate x-axis labels to prevent overlap
    plt.xticks(rotation=45, ha='right', fontsize=10)

    # Add labels to each bar indicating the sentiment
    for bar, sentiment in zip(bars, most_common_sentiment_per_month):
        y_position = bar.get_height() + 10 if bar.get_height() > 10 else bar.get_height() - 5
        plt.text(bar.get_x() + bar.get_width() / 2, y_position, sentiment,
                 ha='center', va='bottom' if bar.get_height() > 10 else 'top', fontsize=10, rotation=90)

    # Show the plot before saving
    plt.tight_layout()

    # Clean the module name for the filename: replace invalid characters with dashes
    safe_module_name = re.sub(r'[^\w\s-]', '', module_name)  # Remove any special characters
    safe_module_name = re.sub(r'[-\s]+', '-', safe_module_name)  # Replace spaces and multiple dashes with a single dash
    safe_topic_name = re.sub(r'[^\w\s-]', '', topic)  # Remove any special characters
    safe_topic_name = re.sub(r'[-\s]+', '-', safe_topic_name)  # Replace spaces and multiple dashes with a single dash

    # Create the filename using the cleaned module name
    plot_filename = f'{output_dir}sentiment_timeseries_{safe_module_name}_{safe_topic_name}.png'

    # Debug statement to print the file path
    print(f"Saving plot to {plot_filename}")

    # Save the plot to a file after rendering
    plt.savefig(plot_filename, dpi=300, bbox_inches='tight', pad_inches=0.1)

    # Display the plot
    plt.show()

    # Close the figure after saving
    plt.close()

    return plot_filename

def plot_time_series_emotion_topic(df: pd.DataFrame, output_dir: str, module_name: str, topic: str) -> str:
    """Generate and show a plot of the most common emotion per month for a given topic."""
    df = df[
        (df['name'] == module_name) &  # Filter by module name
        (df['topic_summary'].astype(str).str.contains(topic, na=False))  # Filter by topic in topic_summary
    ]

    # Ensure the output directory exists
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Group by month and emotion, then count the occurrences of each emotion
    monthly_emotion_counts = full_df_need.groupby(['month', 'emotion_label']).size().unstack(fill_value=0)

    # Find the most common emotion for each month
    most_common_emotion_per_month = monthly_emotion_counts.idxmax(axis=1)

    # Find the count of the most common emotion for each month
    most_common_emotion_count_per_month = monthly_emotion_counts.max(axis=1)

    # Map emotions to their respective numbers for better visualization
    most_common_emotion_numeric = most_common_emotion_per_month.map(emotion_to_number)

    # Create a plot for the most common emotion per month
    plt.figure(figsize=(30, 14))  # Adjusted figure size for better visibility

    # Create a bar plot
    bars = plt.bar(most_common_emotion_per_month.index, most_common_emotion_count_per_month, color='skyblue', width=0.6)

    # Add titles and labels
    plt.title(f'Most Common Emotion for {topic} ({module_name}) (Monthly)', fontsize=14)
    plt.xlabel('Month', fontsize=12)
    plt.ylabel('Number of Reviews', fontsize=12)

    # Rotate x-axis labels to prevent overlap
    plt.xticks(rotation=45, ha='right', fontsize=10)

    # Add labels to each bar indicating the emotion number (mapped value)
    # Add labels to each bar indicating the emotion number (mapped value) with vertical text
    for bar, emotion, numeric_emotion in zip(bars, most_common_emotion_per_month, most_common_emotion_numeric):
        y_position = bar.get_height() + 10 if bar.get_height() > 10 else bar.get_height() - 5
        plt.text(bar.get_x() + bar.get_width() / 2, y_position, f'{emotion} ({numeric_emotion})',
                ha='center', va='bottom' if bar.get_height() > 10 else 'top', fontsize=10, rotation=90)


    # Show the plot before saving
    plt.tight_layout()

    # Clean the module name for the filename: replace invalid characters with dashes
    safe_module_name = re.sub(r'[^\w\s-]', '', module_name)  # Remove any special characters
    safe_module_name = re.sub(r'[-\s]+', '-', safe_module_name)  # Replace spaces and multiple dashes with a single dash
    safe_topic_name = re.sub(r'[^\w\s-]', '', topic)  # Remove any special characters
    safe_topic_name = re.sub(r'[-\s]+', '-', safe_topic_name)  # Replace spaces and multiple dashes with a single dash

    # Create the filename using the cleaned module name
    plot_filename = f'{output_dir}emotion_timeseries_{safe_module_name}_{safe_topic_name}.png'

    # Debug statement to print the file path
    print(f"Saving plot to {plot_filename}")

    # Save the plot to a file after rendering
    plt.savefig(plot_filename, dpi=300, bbox_inches='tight', pad_inches=0.1)

    # Display the plot
    plt.show()

    # Close the figure after saving
    plt.close()

    return plot_filename


def compute_analysis_refs_topic(module_name: str, full_df_need: pd.DataFrame, output_dir: str, topic: str) -> str:
    """Generate analysis reference names for sentiment and emotion time series data for a given topic."""
    # Assuming the module_name is relevant for the time series data plots, call the plot functions
    emotion_plot_file = plot_time_series_emotion_topic(full_df_need, output_dir, module_name, topic)
    sentiment_plot_file = plot_time_series_sentiment_topic(full_df_need, output_dir, module_name, topic)

    # Create a tuple of the filenames (or analysis references)
    analysis_refs = (emotion_plot_file, sentiment_plot_file)

    # Return the file names as a string (or can be returned as a tuple if needed)
    return analysis_refs
```

```python
import base64
import os
from google import genai
from google.genai import types

def generate2(longform_recs: str):
    client = genai.Client(
        api_key="SECRET",
    )

    model = "gemini-2.0-flash"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""i will give the model input to summarise"""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Okay, I'm ready. Please provide the input you want me to summarize. I will do my best to provide a concise and accurate summary based on the information you give me. Just paste it in, and I'll get to work!
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""i want to be able to pass any input string to be summarised"""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Okay, I understand. I'm designed to handle that. You can pass me any string of text, and I will attempt to summarize it. The length and complexity of the text will affect the quality and detail of the summary, but I will do my best in all cases.

Just paste the string you want summarized and I'll get to work! Let me know if you have any specific length constraints or focus areas for the summary as well.
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""the summary should be concise and focus on key areas of improvement listed in the recommendation."""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Got it. So, to reiterate:

1.  **You will provide me with a text string.**
2.  **I will summarize the text, prioritizing brevity.**
3.  **The summary will specifically focus on key areas of improvement mentioned as recommendations within the text.**

Understood. Please paste the text string now. I'm ready when you are.
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""the key areas of improvement should be specified, e.g., instructor and what the actual issue is and the direct recommendation, while maintaining being concise. I like the numbered points as it makes the recommendation clear."""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Okay, I understand. So, now the specific guidelines for the summary are:

1.  **Concise:** Prioritize brevity and clarity.
2.  **Focus on Improvements:** Summarize only the key areas where improvements are recommended.
3.  **Specificity:** Identify the subject of the improvement (e.g., \"Instructor: [Instructor's Name]\") and the specific issue.
4.  **Direct Recommendation:** State the direct recommendation as it appears in the text.
5.  **Numbered Points:** Present the recommendations in a numbered list format for clarity.

This is very helpful.  I'm ready to receive the input text!
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text=longform_recs
                ),
            ],
        ),
    ]
    generate_content_config = types.GenerateContentConfig(
        temperature=1,
        top_p=0.95,
        top_k=40,
        max_output_tokens=8192,
        response_mime_type="text/plain",
    )

    # Collect the response
    summary = ""
    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        summary += chunk.text
        print(chunk.text, end="")

    return summary

```

```python
import base64
import os
from google import genai
from google.genai import types

def generate2(longform_recs: str):
    client = genai.Client(
        api_key="SECRET",
    )

    model = "gemini-2.0-flash"
    contents = [
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""i will give the model input to summarise"""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Okay, I'm ready. Please provide the input you want me to summarize. I will do my best to provide a concise and accurate summary based on the information you give me. Just paste it in, and I'll get to work!
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""i want to be able to pass any input string to be summarised"""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Okay, I understand. I'm designed to handle that. You can pass me any string of text, and I will attempt to summarize it. The length and complexity of the text will affect the quality and detail of the summary, but I will do my best in all cases.

Just paste the string you want summarized and I'll get to work! Let me know if you have any specific length constraints or focus areas for the summary as well.
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""the summary should be concise and focus on key areas of improvement listed in the recommendation."""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Got it. So, to reiterate:

1.  **You will provide me with a text string.**
2.  **I will summarize the text, prioritizing brevity.**
3.  **The summary will specifically focus on key areas of improvement mentioned as recommendations within the text.**

Understood. Please paste the text string now. I'm ready when you are.
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text="""the key areas of improvement should be specified, e.g., instructor and what the actual issue is and the direct recommendation, while maintaining being concise. I like the numbered points as it makes the recommendation clear."""
                ),
            ],
        ),
        types.Content(
            role="model",
            parts=[
                types.Part.from_text(
                    text="""Okay, I understand. So, now the specific guidelines for the summary are:

1.  **Concise:** Prioritize brevity and clarity.
2.  **Focus on Improvements:** Summarize only the key areas where improvements are recommended.
3.  **Specificity:** Identify the subject of the improvement (e.g., \"Instructor: [Instructor's Name]\") and the specific issue.
4.  **Direct Recommendation:** State the direct recommendation as it appears in the text.
5.  **Numbered Points:** Present the recommendations in a numbered list format for clarity.

This is very helpful.  I'm ready to receive the input text!
"""
                ),
            ],
        ),
        types.Content(
            role="user",
            parts=[
                types.Part.from_text(
                    text=longform_recs
                ),
            ],
        ),
    ]
    generate_content_config = types.GenerateContentConfig(
        temperature=1,
        top_p=0.95,
        top_k=40,
        max_output_tokens=8192,
        response_mime_type="text/plain",
    )

    # Collect the response
    summary = ""
    for chunk in client.models.generate_content_stream(
        model=model,
        contents=contents,
        config=generate_content_config,
    ):
        summary += chunk.text
        print(chunk.text, end="")

    return summary

```
